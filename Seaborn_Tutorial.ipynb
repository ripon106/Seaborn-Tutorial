```python
import pandas as pd
import seaborn as sbs
import matplotlib.pyplot as plt
```

# New Section

```python
df = pd.read_csv('/content/seaborn.csv')
df.head()
```

```python
x = df['qsec'].values
y = df['carb'].values
```

Scatter, plot

```python
plt.figure(figsize=(7,5));#figure size
plt.scatter(x,y)# dot print
plt.plot(x,y)#line
plt.show()#show 
```

Pairplot


```python
sbs.pairplot(df, height = 1.5)
```

```python
df.head(2)
```

print pariplot to take some element

```python
key = ['wt','qsec','gear','carb']
sbs.pairplot(df[key], height = 1.5)
plt.show()
```

Heatmap


```python
sbs.heatmap(df[key].corr(), annot = True,fmt=".2f")
plt.show()
```

Join plot

```python
sbs.jointplot(x='gear', y='carb', data=df, kind='reg', height=8);
plt.show();
```

Regression plot

```python
sbs.regplot(x = 'gear', y = 'carb', data = df);
```
